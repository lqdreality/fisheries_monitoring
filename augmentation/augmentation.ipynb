```python
import numpy as np
from scipy.misc import imread
from imgaug import augmenters
import matplotlib.pyplot as plt
```

```python
# Setup some augmenters
horz_flip = augmenters.Fliplr(1.0)
blur = augmenters.GaussianBlur(3.0)
```

```python
# Apply these augmenters to single images
#img = np.array(imread('/a/data/fisheries_monitoring/cropped_data/YFT/img_01539-0.jpg'))
img = np.array(imread('/a/data/fisheries_monitoring/data/train/YFT/img_02121.jpg'))
flipped_img = horz_flip.augment_image(img)
blurred_img = blur.augment_image(img)
fig = plt.figure(figsize=(25, 25))
ax = fig.add_subplot(1,3,1)
plt.imshow(img)
ax.set_title('Original')
ax = fig.add_subplot(1,3,2)
plt.imshow(flipped_img)
ax.set_title('Horizontal Flip')
ax = fig.add_subplot(1,3,3)
plt.imshow(blurred_img)
ax.set_title('Blurred')
plt.show()
```
