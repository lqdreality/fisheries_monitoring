## Create crops of the No Fish Images

```python
import numpy as np
import glob
from scipy.misc import imread
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
# Load the data
os.chdir('/a/data/fisheries_monitoring/data/train/NoF')
train_files = sorted(glob.glob('*.jpg'))
```

```python
# Generate random BB's based off of the distribution of BB sizes from the original data
#!awk '{ print $3","$4 }' FS="," /a/data/fisheries_monitoring/data/detection_labels/boxes.csv > /tmp/bb.csv
width_height = np.genfromtxt('/tmp/bb.csv', delimiter=',')
widths = width_height[:,0]
heights = width_height[:,1]
counts = plt.hist2d(widths, heights, bins=100)
plt.plot(np.array(range(0,600)), np.sqrt(600**2 - np.array(range(0,600))**2), 'r.')
plt.plot(np.array(range(0,800)), np.array(range(0,800)),'r-')
plt.xlabel('Widths')
plt.ylabel('Heights')
plt.show()
```

```python
# Region of 200-500 Widths, 50-250 heights & 50-150 Widths, 300-400 Heights
for file_name in train_files :
    print 'Processing ' + file_name
    img = Image.open(file_name)
    W,H = img.size
    # Generate only one box per image (can increase this later by adding a loop)
    if np.random.randint(0,100) < 35 :
        width = np.random.randint(50,150)
        height = np.random.randint(300,400)
    else :
        width = np.random.randint(200,500)
        height = np.random.randint(50,250)
    left = np.random.randint(0,W-width)
    right = left + width
    top = np.random.randint(0,H-height)
    bottom = top + height
    cropped_img = img.crop((left, top, right, bottom))
    cropped_img.save('/a/data/fisheries_monitoring/cropped_data/NoF/' + file_name[:-4] + '-1.jpg')
    #plt.imshow(img)
    #plt.show()
    #plt.imshow(cropped_img)
    #plt.show()
```
